```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName("pyspark_select").getOrCreate()
```

```python
data = [("James","Smith","USA","CA"),
    ("Michael","Rose","USA","NY"),
    ("Robert","Williams","USA","CA"),
    ("Maria","Jones","USA","FL")]
column = ["fname","lname","country","state"]
df = spark.createDataFrame(data,column)
df.printSchema()
```

```python
df.select([col for col in df.columns]).show()
df.select("*").show()
```
