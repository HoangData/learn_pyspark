```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col
import pyspark
```

```python
spark = SparkSession.builder.appName("pyspark_distinct").getOrCreate()
```

```python
data = [("James", "Sales", 3000), \
    ("Michael", "Sales", 4600), \
    ("Robert", "Sales", 4100), \
    ("Maria", "Finance", 3000), \
    ("James", "Sales", 3000), \
    ("Scott", "Finance", 3300), \
    ("Jen", "Finance", 3900), \
    ("Jeff", "Marketing", 3000), \
    ("Kumar", "Marketing", 2000), \
    ("Saif", "Sales", 4100) \
  ]
columns= ["employee_name", "department", "salary"]
df = spark.createDataFrame(data = data, schema = columns)
df.printSchema()
```

**Total number of records after removing the duplicate records**

```python
print("C1: " + str(df.distinct().count()))
print("C2: " + str(df.dropDuplicates().count()))
```

**Distinct of select multiple columns**

```python
dropDuplicate = df.dropDuplicates(['department','salary'])
print(dropDuplicate.count())
dropDuplicate.show()
```
